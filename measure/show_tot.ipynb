```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('tot_pc_screen_4/tot_pc_screen_4.json', 'r') as file:
    data = json.load( file )
frames = np.asarray( data['frames'] )
```

```python
num_frames = frames.shape[0]
num_cum = 10000

idxs = np.arange(0, num_frames, num_cum)
print( frames.shape )
```

```python
for idx in range(len(idxs) - 1):
    hist, bins = np.histogram(frames[idxs[idx]:idxs[idx+1]].flatten(), bins=np.arange(1, 400))
    plt.step(bins[:-1], hist)
    # plt.show()
```
